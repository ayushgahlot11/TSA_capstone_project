```python
# Task 2: Data Preprocessing and Stationarity Testing
# Author: Ayush Gahlot

import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import adfuller
from sklearn.preprocessing import MinMaxScaler
import os
import warnings
warnings.filterwarnings('ignore')

# Make sure the processed data folder exists
os.makedirs('../data/processed', exist_ok=True)

# Load the raw dataset we just saved in Task 1
print("Loading raw dataset...")
data = pd.read_csv('../data/raw/capstone_raw_data.csv', index_col=0, parse_dates=True)
print(f"Loaded data with {data.shape[1]} stocks.\n")

# --- 1. Stationarity Testing (ADF Test) ---
# Time series models like ARIMA need stationary data to work properly

def check_stationarity(series):
    # Running the Augmented Dickey-Fuller test
    result = adfuller(series.dropna())
    p_value = result[1]
    return p_value, p_value < 0.05 # Returns True if it is stationary

adf_results = []
differenced_data = pd.DataFrame(index=data.index)

print("Running ADF Tests for Stationarity...")
for col in data.columns:
    p_val, is_stationary = check_stationarity(data[col])
    
    if is_stationary:
        status = "Stationary"
        differenced_data[col] = data[col] # Keep as is
    else:
        status = "Non-Stationary (Differencing Applied)"
        differenced_data[col] = data[col].diff() # Apply 1st order differencing
        
    adf_results.append({'Stock': col, 'P-Value': round(p_val, 4), 'Status': status})

# Print out a clean table showing what we differenced
adf_df = pd.DataFrame(adf_results)
print(adf_df.to_string(index=False))
print("\nFirst-order differencing applied where needed.\n")

# --- 2. Data Scaling ---
# Scaling prices between 0 and 1 so neural networks (like LSTM) don't freak out
print("Scaling data using MinMaxScaler...")
scaler = MinMaxScaler(feature_range=(0, 1))

# Drop the first row since differencing creates a NaN at the very beginning
differenced_data.dropna(inplace=True)
scaled_values = scaler.fit_transform(differenced_data)

scaled_data = pd.DataFrame(scaled_values, columns=differenced_data.columns, index=differenced_data.index)

# --- 3. Train / Test Split ---
# The rubric explicitly says to use July 2025 to Dec 2025 as the test set
split_date = '2025-07-01'

train_set = scaled_data[:split_date]
test_set = scaled_data[split_date:]

print(f"Train/Test Split Done:")
print(f"Training Data: {train_set.shape[0]} days")
print(f"Testing Data: {test_set.shape[0]} days\n")

# Save the clean, scaled, and split data for Task 3
train_set.to_csv('../data/processed/train_data.csv')
test_set.to_csv('../data/processed/test_data.csv')
print("Task 2 complete. Preprocessed data saved to 'data/processed/' folder.")
```
