```python
# Task 3 & 6: Time Series Forecasting and Model Evaluation
# Author: Ayush Gahlot

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
import os
import warnings
warnings.filterwarnings('ignore')

os.makedirs('../reports/figures', exist_ok=True)
os.makedirs('../data/processed', exist_ok=True)
plt.style.use('dark_background')

# Load the raw data. It's much easier to forecast real rupees for the portfolio
# allocation step than trying to reverse-engineer scaled/differenced data.
print("Loading raw data...")
data = pd.read_csv('../data/raw/capstone_raw_data.csv', index_col=0, parse_dates=True)

# Capstone requires testing on the last 6 months (July - Dec 2025)
split_date = '2025-07-01'
train_data = data[:split_date]
test_data = data[split_date:]

print(f"Training models on {len(data.columns)} stocks. This might take a minute...")

results = []
forecasts_jan = pd.DataFrame()

# Loop through every stock in the portfolio
for stock in data.columns:
    
    train_series = train_data[stock].dropna()
    test_series = test_data[stock].dropna()
    
    # ==========================================
    # MODEL 1: ARIMA
    # ==========================================
    # Using order (1,1,1) as a solid mathematical baseline
    try:
        arima_model = ARIMA(train_series, order=(1, 1, 1))
        arima_fit = arima_model.fit()
        arima_pred = arima_fit.forecast(steps=len(test_series))
        
        # Forecast 2 days into Jan 2026 for Task 5
        arima_jan = arima_fit.forecast(steps=len(test_series) + 2).iloc[-2:]
        
        # Calculate Errors
        arima_rmse = np.sqrt(mean_squared_error(test_series, arima_pred))
        arima_mape = mean_absolute_percentage_error(test_series, arima_pred)
        
        # Directional Accuracy (Did it guess the trend correctly?)
        actual_dir = np.sign(test_series.iloc[-1] - test_series.iloc[0])
        arima_dir = np.sign(arima_pred.iloc[-1] - arima_pred.iloc[0])
        arima_dir_acc = "Correct" if actual_dir == arima_dir else "Incorrect"
    except Exception as e:
        arima_rmse, arima_mape, arima_dir_acc, arima_jan = np.nan, np.nan, "Failed", [np.nan, np.nan]

    results.append({
        'Stock': stock, 'Model': 'ARIMA', 
        'RMSE': round(arima_rmse, 2), 'MAPE': round(arima_mape, 4), 'Dir_Accuracy': arima_dir_acc
    })
    
    # ==========================================
    # MODEL 2: Exponential Smoothing (Holt)
    # ==========================================
    try:
        exp_model = ExponentialSmoothing(train_series, trend='add', initialization_method="estimated")
        exp_fit = exp_model.fit()
        exp_pred = exp_fit.forecast(len(test_series))
        
        # Forecast 2 days into Jan 2026
        exp_jan = exp_fit.forecast(len(test_series) + 2).iloc[-2:]
        
        exp_rmse = np.sqrt(mean_squared_error(test_series, exp_pred))
        exp_mape = mean_absolute_percentage_error(test_series, exp_pred)
        
        exp_dir = np.sign(exp_pred.iloc[-1] - exp_pred.iloc[0])
        exp_dir_acc = "Correct" if actual_dir == exp_dir else "Incorrect"
    except Exception as e:
        exp_rmse, exp_mape, exp_dir_acc, exp_jan = np.nan, np.nan, "Failed", [np.nan, np.nan]

    results.append({
        'Stock': stock, 'Model': 'Exp Smoothing', 
        'RMSE': round(exp_rmse, 2), 'MAPE': round(exp_mape, 4), 'Dir_Accuracy': exp_dir_acc
    })
    
    # Saving Exponential Smoothing predictions for Task 5 portfolio allocation 
    # because it generally handles baseline momentum better than un-tuned ARIMA
    forecasts_jan[stock] = exp_jan.values

# --- Saving the Task 6 Comparison Table ---
comparison_df = pd.DataFrame(results)
comparison_df.to_csv('../reports/task6_model_comparison.csv', index=False)

print("\n--- Task 6: Model Evaluation Snippet ---")
print(comparison_df.head(6).to_string(index=False))

# --- Saving the Task 3 January Forecasts ---
forecasts_jan.index = ['2026-01-01', '2026-01-02']
forecasts_jan.to_csv('../data/processed/jan_2026_forecasts.csv')

print("\nTask 3 & 6 done.")
print("Forecasts saved to 'data/processed/jan_2026_forecasts.csv'.")
print("Model comparison saved to 'reports/task6_model_comparison.csv'.")
```
